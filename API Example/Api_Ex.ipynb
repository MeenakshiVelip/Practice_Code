```python
import requests

def get_github_user(username):
    base_url = f'https://api.github.com/users/{username}'
    
    # Sending a GET request to the GitHub API to fetch user data
    response = requests.get(base_url)
    
    if response.status_code == 200:
        # Request successful, parse the JSON response
        user_data = response.json()
        return user_data
    else:
        # Request failed
        print(f"Failed to fetch user data. Status code: {response.status_code}")
        return None

def get_user_repositories(username):
    base_url = f'https://api.github.com/users/{username}/repos'
    
    # Sending a GET request to the GitHub API to fetch user repositories
    response = requests.get(base_url)
    
    if response.status_code == 200:
        # Request successful, parse the JSON response
        repositories_data = response.json()
        return repositories_data
    else:
        # Request failed
        print(f"Failed to fetch repositories. Status code: {response.status_code}")
        return None

# Replace 'YOUR_GITHUB_USERNAME' with the GitHub username you want to retrieve information for
github_username = 'MeenakshiVelip'

# Fetch user information
user_info = get_github_user(github_username)

if user_info:
    print(f"GitHub User: {user_info['login']}")
    print(f"Name: {user_info['name']}")
    print(f"Followers: {user_info['followers']}")
    print(f"Following: {user_info['following']}")
    print(f"Public Repositories: {user_info['public_repos']}")

    # Fetch user repositories
    repositories = get_user_repositories(github_username)
    
    if repositories:
        print("\nRepositories:")
        for repo in repositories:
            print(f"- {repo['name']}: {repo['html_url']}")
    else:
        print("No repositories found.")
else:
    print("User data retrieval failed.")
```
